```python
import pandas as pd
import seaborn as sns
import ptitprince as pt
from scipy.io import loadmat
import matplotlib.pyplot as plt
# pandas==1.5.3
# seaborn==0.11
# ptitprince==0.2.7
# matplotlib==3.8.4

n_qudits = 4
time_str = '20241008_155835'
path = f'./mats/VGON_nqd{n_qudits}_{time_str}'
overlaps = loadmat(f'{path}.mat')['overlaps']

data = {
    r'$|v_1\rangle$': overlaps[:, 0],
    r'$|v_2\rangle$': overlaps[:, 1],
    r'$|v_3\rangle$': overlaps[:, 2],
    r'$|v_4\rangle$': overlaps[:, 3],
}
colors = ['#6FAE45', '#589CD6', '#4474C4', '#660874', '#AB2A3C']
df = pd.DataFrame(data)

fig, ax = plt.subplots()
ax = pt.half_violinplot(data=df, inner=None, width=1, palette=colors, color='white')
ax = sns.stripplot(data=df, jitter=True, palette=colors, size=2, edgecolor='none')
plt.show()
```
